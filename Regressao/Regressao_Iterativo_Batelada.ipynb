```python
#Regressão linear: Estimação de preços de apartamentos a partir do número de banheiros que eles possuem.

#obs: Foi utilizado o Método do Gradiente Descendente - Batelada


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('kc_house_data.csv')
df = df.loc[df.bedrooms < 33]
df = df.groupby(['bedrooms']).mean()
new_df = df.groupby(['bedrooms']).mean()
x = new_df.index.values
y = new_df.price.values
```

```python
temp = np.ones((len(x), 1), dtype = int)
x = x.reshape(len(x), 1)
y = y.reshape(len(y), 1)
fi = np.concatenate((temp, x), axis = 1)
```

```python
def grad(w, fi, y):
        w_t = w.transpose()
        fi_t = fi.transpose()
        y_t = y.transpose()
        e_t = np.matmul(w_t, fi_t) - y_t
        grad_Je = (2) * (np.matmul(e_t, fi))
        grad_Je = grad_Je.transpose()
        return grad_Je
        
```

```python
w = np.random.rand(2)
w = w.reshape(2,1)
alfa = .001
W = w

for i in range(1000):
    w  = w - alfa * grad(w, fi, y)
    W = np.hstack((W, w))
```

```python
x_n = np.arange(min(x), max(x), 1)
y_n = np.array([])
for i in x_n:
    y_n = np.append(y_n, [w[0]+w[1]*i])
```

```python
fig, ax = plt.subplots()
ax.plot(x_n, y_n)
ax.stem(x, y)
ax.set_ylabel('Preço ($)')
ax.set_xlabel('nº de banheiros')
plt.show()
```

```python
w0 = W[0, :]
w1 = W[1, :]
fig, bx = plt.subplots()
bx.plot(w0, w1)
bx.set_ylabel('w1')
bx.set_xlabel('w0')
plt.show()
```
