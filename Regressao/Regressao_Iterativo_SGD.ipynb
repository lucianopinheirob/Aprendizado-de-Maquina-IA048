```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('kc_house_data.csv')
df = df.loc[df.bedrooms < 33]
df = df.groupby(['bedrooms']).mean()
new_df = df.groupby(['bedrooms']).mean()
x = new_df.index.values
y = new_df.price.values
```

```python
temp = np.ones((len(x), 1), dtype = int)
x = x.reshape(len(x), 1)
y = y.reshape(len(y), 1)
fi = np.concatenate((temp, x), axis = 1)
```

```python
w = np.random.rand(2)
w = w.reshape(2,1)
W = w
alfa = .001
epocas = 5000

for i in range(epocas):
    for y_linha, fi_linha in zip(y, fi):
        grad = -2 * (y_linha - np.matmul(fi_linha, w)) * fi_linha.transpose()
        grad = grad.reshape(2,1)
        w = w - alfa*grad
        W = np.hstack((W, w))
```

```python
x_n = np.arange(min(x), max(x), 1)
y_n = np.array([])
for i in x_n:
    y_n = np.append(y_n, [w[0]+w[1]*i])
```

```python
fig, ax = plt.subplots()
ax.plot(x_n, y_n)
ax.stem(x, y)
ax.set_ylabel('Preço ($)')
ax.set_xlabel('nº de banheiros')
plt.show()
```

```python
w0 = W[0, :]
w1 = W[1, :]
fig, bx = plt.subplots()
bx.plot(w0, w1)
bx.set_ylabel('w1')
bx.set_xlabel('w0')
plt.show()
```
