Para a atividade foram utilizadas algumas funções prontas da biblioteca ScikitLearn, já que algumas delas demandariam uma complexidade razoavelmente grande para serem feitas do zero.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

A função Formata_Matriz foi utilizada para formatar a matriz de confusão com o nome das células dos bancos de imagem.

```python
def Formata_Matriz(mc):
    
    Colunas = {"Basófilos": mc[:, 0],
    "Eosinófilos": mc[:, 1] ,
    "Eritroblastos": mc[:, 2],
    "Granulócitos Imaturos": mc[:, 3],
    "Linfócitos": mc[:, 4],
    "Monócitos": mc[:, 5],
    "Neutrófilos": mc[:, 6],
    "Plaquetas": mc[:, 7]}
    Indice = ["Basófilos", "Eosinófilos", "Eritroblastos", "Granulócitos Imaturos", "Linfócitos", "Monócitos", "Neutrófilos", "Plaquetas"]
    mc_df = pd.DataFrame(Colunas, index = Indice)
    
    return mc_df
```

Os conjuntos de imagens de treino, teste e validação foram carregados e normalizados. Em seguida as imagens, que são compostas de 3 canais de cor, foram redimensionadas com o objetivo de servirem como entrada para a rede neural MLP.

```python
Train_Images = np.load('train_images.npy')/255
Train_Labels = np.load('train_labels.npy')

Val_Images = np.load('val_images.npy')/255
Val_Labels = np.load('val_labels.npy')

Test_Images = np.load('test_images.npy')/255
Test_Labels = np.load('test_labels.npy')

Train_Images = np.reshape(Train_Images, (len(Train_Images), 28*28*3))
Train_Labels = np.reshape(Train_Labels, len(Train_Labels))

Val_Images = np.reshape(Val_Images, (len(Val_Images), 28*28*3))
Val_Labels = np.reshape(Val_Labels, len(Val_Labels))

Test_Images = np.reshape(Test_Images, (len(Test_Images), 28*28*3))
Test_Labels = np.reshape(Test_Labels, len(Test_Labels))
```

A primeira rede neural consistiu de apenas uma camada intermediária e foi configurada da seguinte maneira: o número de neurônios da camada foi aumentado gradualmente a partir de 5, até que se consiguisse obter os melhores resultados na etapa de validação, o que foi alcançado com 8 neurônios. No decorrer do processo foi importante realizar um ajuste fino da tolerância para o critério de parada e a taxa de aprendizado para que o modelo conseguisse ter um desempenho melhor. O algoritmo utilizado para a otimização foi o "Adam" e a função de ativação a "Relu", que proporcionaram os melhores resultados e a taxa de aprendizado foi mantida constante durante todo o procedimento, por simplicidade.

```python
Neuronio_1 = MLPClassifier(hidden_layer_sizes = (8),
                   max_iter = 600,
                   tol = 0.0000001,
                   learning_rate_init = .0001,
                   solver = "adam",
                   activation = "relu",
                   learning_rate = "constant",
                   verbose = 2
                  )

Neuronio_1.fit(Train_Images, Train_Labels)
```

A segunda rede neural foi configurada de maneira semelhante a primeira. Primeiramente foram tentadas duas camadas de oito neuronios. Após isso os valores de neurônios foram sendo incrementados de um em um até que se obtesse a configuração que gerou o melhor resultado nos dados de validação, com 10 neurônios na primeira camada e 10 neurônios na segunda. É importante pontuar que por estar sendo usada uma biblioteca pronta, houve pouca flexibilidade para determinar os pesos iniciais de treinamento. O modelo, apesar de conseguir resultados melhores, se mostrou mais sensível à inicialização dos parâmetros.

```python
Neuronio_2 = MLPClassifier(hidden_layer_sizes = (10, 10),
                   max_iter = 600,
                   tol = 0.0000001,
                   learning_rate_init = .0001,
                   solver = "adam",
                   activation = "relu",
                   learning_rate = "constant",
                   verbose = 2,
                   alpha = 0.00001
                  )

Neuronio_2.fit(Train_Images, Train_Labels)

```

```python
#Predição das Imagens de Validação
Pred_1 = Neuronio_1.predict(Test_Images)

#Acurácia
print("Acurácia MLP 1 camada:", accuracy_score(Test_Labels, Pred_1))
```

```python
#Matriz de Confusão
mc_1 = confusion_matrix(Test_Labels, Pred_1, labels=[0, 1, 2, 3, 4, 5, 6, 7])
print("Matriz de confusão MLP 1 camada")
Formata_Matriz(mc_1)
```

```python
#Predição das Imagens de Validação
Pred_2 = Neuronio_2.predict(Test_Images)

#Acurácia
print("Acurácia MLP 2 camadas:", accuracy_score(Test_Labels, Pred_2))
```

```python
#Matriz de Confusão
mc_2 = confusion_matrix(Test_Labels, Pred_2, labels=[0, 1, 2, 3, 4, 5, 6, 7])
print("Matriz de confusão MLP duas camadas:")
Formata_Matriz(mc_2)
```

Houve um aumento de aproximadamente 11% na acurácia na segunda rede neural em relação a primeira. Observando a matriz de confusão, fica evidente que houve um aumento expressivo no acerto em relação às classes Basófilos - 0 na primeira e 144 na segunda - e Monócitos - 0 na primeira e 112 na segunda.
