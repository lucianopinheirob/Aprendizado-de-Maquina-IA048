```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.set_printoptions(precision=10)

Train_images = np.load('train_images.npy')
Train_labels = np.load('train_labels.npy')

Val_images = np.load('val_images.npy')
Val_labels = np.load('val_labels.npy')

Test_images = np.load('test_images.npy')
Test_labels = np.load('test_labels.npy')

uns = 0
zeros = 0
for Val_label in Val_labels:
    if Val_label == 1:
        uns += 1
    else:
        zeros += 1
```

```python
#Formatação dos dados com base no valor absoluto dos autovalores das matrizes

dim_image = len(Train_images[0])

#Treinamento
Atributos_Train_Empilhados = np.empty((0, 88))
for image in Train_images:
    Soma = np.sum(image) / (dim_image ** 2)
    Traco = np.trace(image)
    Autovalores = np.linalg.eig(image)[0]
    Autovalores = np.absolute(Autovalores)
    Soma_Lin = np.array([0])
    Soma_Col = np.array([0])
    for lin in image:
        Soma_Lin = np.hstack((Soma_Lin, np.sum(lin) / dim_image))
    for col in image.T:
        Soma_Col = np.hstack((Soma_Col, np.sum(col) / dim_image))
    Atributos_Train = np.hstack((Soma, Autovalores, Traco, Soma_Lin, Soma_Col))
    Atributos_Train_Empilhados = np.vstack((Atributos_Train_Empilhados, Atributos_Train)) 
Fi_Train = np.hstack((np.ones((len(Atributos_Train_Empilhados), 1)), Atributos_Train_Empilhados))

#Validacao
Atributos_Val_Empilhados = np.empty((0, 88))
for image in Val_images:
    Soma = np.sum(image) / (dim_image ** 2)
    Traco = np.trace(image)
    Autovalores = np.linalg.eig(image)[0]
    Autovalores = np.absolute(Autovalores)
    Soma_Lin = np.array([0])
    Soma_Col = np.array([0])
    for lin in image:
        Soma_Lin = np.hstack((Soma_Lin, np.sum(lin) / dim_image))
    for col in image.T:
        Soma_Col = np.hstack((Soma_Col, np.sum(col) / dim_image))
    Atributos_Val = np.hstack((Soma, Autovalores, Traco, Soma_Lin, Soma_Col))
    Atributos_Val_Empilhados = np.vstack((Atributos_Val_Empilhados, Atributos_Val)) 
Fi_Val = np.hstack((np.ones((len(Atributos_Val_Empilhados), 1)), Atributos_Val_Empilhados))

#Teste
Atributos_Test_Empilhados = np.empty((0, 88))
for image in Test_images:
    Soma = np.sum(image) / (dim_image ** 2)
    Traco = np.trace(image)
    Autovalores = np.linalg.eig(image)[0]
    Autovalores = np.absolute(Autovalores)
    Soma_Lin = np.array([0])
    Soma_Col = np.array([0])
    for lin in image:
        Soma_Lin = np.hstack((Soma_Lin, np.sum(lin) / dim_image))
    for col in image.T:
        Soma_Col = np.hstack((Soma_Col, np.sum(col) / dim_image))
    Atributos_Test = np.hstack((Soma, Autovalores, Traco, Soma_Lin, Soma_Col))
    Atributos_Test_Empilhados = np.vstack((Atributos_Test_Empilhados, Atributos_Test)) 
Fi_Test = np.hstack((np.ones((len(Atributos_Test_Empilhados), 1)), Atributos_Test_Empilhados))
```

```python
#Treinamento
def grad(Fi_Train, w, Train_labels):
    e = np.empty((0, 1))
    for i in range(len(Fi_Train)):
        label_calc = 1 / (1 + np.exp(-(np.dot(Fi_Train[i], w))))
        erro = Train_labels[i][0] - label_calc
        e = np.vstack((e, erro))
    grad = - 1/len(Fi_Train) * (np.matmul(e.T, Fi_Train))
    return grad[0]

w = np.random.rand(len(Fi_Train[0])) #Pode ser negativo tbm, ver depois

#a = [.5]
#w = np.array(len(Fi_Train[0])*a)

alfa = .0001
for i in range(10000):
    w  = w - alfa * grad(Fi_Train, w, Train_labels)
    if i % 1000 == 0: 
        print(i)
```

```python
#Validacao
Label_calcs = np.empty((0,1))
for i in range(len(Fi_Val)):
    Label_calc = 1 / (1 + np.exp(-(np.dot(Fi_Val[i], w))))
    if Label_calc >= 0.5:
        Label_calc = 1
    else:
        Label_calc = 0
    Label_calcs = np.vstack((Label_calcs, Label_calc))
    


Conta_Tipos = {(1, 1): 0, (1, 0): 0, (0, 1): 0, (0, 0):0}

for Label_calc, Val_label in zip(Label_calcs, Val_labels):    
    Conta_Tipos[(Label_calc[0], Val_label[0])] += 1
    
TP = Conta_Tipos[(1,1)]
FP = Conta_Tipos[(1,0)]
FN = Conta_Tipos[(0,1)]
TN = Conta_Tipos[(0,0)]
    

Matriz_Conf = np.array([[TP, FN], 
                        [FP, TN]])    
    
Espe = TN / (TN + FP)
Sens = TP / (TP + FN)
Prec = TP / (TP + FP)
TFPo = FP / (TN + FP)

BA = (Espe + Sens) / 2
Acc = (TP + TN) / (TP + TN + FP + FN)
Fm = (2 * Sens * Prec) / (Sens + Prec)

```

```python
#Teste
Label_calcs = np.empty((0,1))
Label_calcs_Bruto = np.empty((0,1)) #Será usado para curva ROC
for i in range(len(Fi_Test)):
    Label_calc = 1 / (1 + np.exp(-(np.dot(Fi_Test[i], w))))
    Label_calcs_Bruto = np.vstack((Label_calcs_Bruto, Label_calc))
    if Label_calc >= 0.5:
        Label_calc = 1
    else:
        Label_calc = 0
    Label_calcs = np.vstack((Label_calcs, Label_calc))
    


Conta_Tipos = {(1, 1): 0, (1, 0): 0, (0, 1): 0, (0, 0):0}

for Label_calc, Test_label in zip(Label_calcs, Test_labels):    
    Conta_Tipos[(Label_calc[0], Test_label[0])] += 1
    
TP = Conta_Tipos[(1,1)]
FP = Conta_Tipos[(1,0)]
FN = Conta_Tipos[(0,1)]
TN = Conta_Tipos[(0,0)]
    

Matriz_Conf = np.array([[TP, FN], 
                        [FP, TN]])    
    
Espe = TN / (TN + FP)
Sens = TP / (TP + FN)
Prec = TP / (TP + FP)
TFPo = FP / (TN + FP)

BA = (Espe + Sens) / 2
Acc = (TP + TN) / (TP + TN + FP + FN)
Fm = (2 * Sens * Prec) / (Sens + Prec)
```

```python
Acc
```

```python
#Curva ROC

Sens = []
TFPo = []
Thresholds = np.arange(0, 1, 0.00001)

for Threshold in Thresholds:
    
    Label_Calc_ROC = np.empty((0,1))
    for Label_calc_Bruto in Label_calcs_Bruto:
        if Label_calc_Bruto >= Threshold:
            Label_calc_Bruto = 1
        else:
            Label_calc_Bruto = 0
        Label_Calc_ROC = np.vstack((Label_Calc_ROC, Label_calc_Bruto))
    #print(Label_Calc_ROC)

    Conta_Tipos = {(1, 1): 0, (1, 0): 0, (0, 1): 0, (0, 0):0}

    for Label_calc, Test_label in zip(Label_Calc_ROC, Test_labels):    
        Conta_Tipos[(Label_calc[0], Test_label[0])] += 1

    TP = Conta_Tipos[(1,1)]
    FP = Conta_Tipos[(1,0)]
    FN = Conta_Tipos[(0,1)]
    TN = Conta_Tipos[(0,0)]  
    
    #print(TP, FP, FN, TN)

    Sens.append(TP / (TP + FN)) 
    TFPo.append(FP / (FP + TN))
```

```python
fig, ax = plt.subplots()
ax.plot(TFPo, Sens)
ax.set_ylabel('Sens')
ax.set_xlabel('TFPo')
plt.show()
```
